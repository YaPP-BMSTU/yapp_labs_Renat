```python
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
import warnings 
```

```python
fortnite_state = pd.read_csv('Fortnite Statistics.csv')
```

```python
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
```

```python
lineplot = sb.lineplot(fortnite_state, x = 'Head Shots', y = 'Hits').set_title('Линейная диаграмма, которая показывает отношение всех хитов к хэдшотам\n')
xlabel = plt.xlabel('\nПопадания в голову')
ylabel = plt.ylabel('Попаданий всего\n')
```

## Смысл
*График хитов всего к хитам в голову*
## Что было сделано
- Построен линейный график с помощью библеотек seaborn, matplotlib
- На оси X расположил количество попаданий в голову
- На оси Y расположил количество попаданий всего
## Выводы
*По графику можно сделать несколько выводов*
- Чем больше обычных хитов, тем больше хитов в голову
- Конкретно на данном графике нет ситуаций, когда количество хитов всего = хитам в голову
- На попадания в голову сказываются множество других факторов. Например точность игрока 

___

```python
fortnite_state['Accuracy'] = fortnite_state['Accuracy'].apply(lambda x: float(x.rstrip('%')) if isinstance(x, str) else x)
x_tick_step = 10
lineplot = sb.lineplot(data=fortnite_state, x='Accuracy', y='Head Shots')
plt.title('Линейная диаграмма, которая показывает отношение точности к хэдшотам\n')
plt.xlabel('\nТочность')
plt.ylabel('Попаданий в голову\n')
plt.xticks(range(int(min(fortnite_state['Accuracy'])), int(max(fortnite_state['Accuracy'])), x_tick_step))
plt.show()
```

## Смысл
*График отношения точности к хэдшотам*
## Что было сделано
- Построен линейный график с помощью библеотек seaborn, matplotlib
- На оси X расположил точность игроков с шагом в 10
- На оси Y расположил количество попаданий в голову
## Выводы
*По графику можно сделать несколько выводов*

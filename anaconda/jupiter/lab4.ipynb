```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
x_col, y_col = 0, 0
flag = True
while flag:
    try:
        # cols: npg - 2; birth_rate - 3; death_rate - 4; gdw - 5; urbanization - 6
        x_col = int(input("cols: npg - 2; birth_rate - 3; death_rate - 4; gdw - 5; urbanization - 6\n"
                          "Hi, choose X line for linear regression: "))
        y_col = int(input("cols: npg - 2; birth_rate - 3; death_rate - 4; gdw - 5; urbanization - 6\n"
                          "Hi, choose Y line for linear regression: "))
    except:
        x_col = -1
        y_col = -1
    if 1 < x_col < 7 and 1 < y_col < 7:
        flag = False
    else:
        print(" Please try again! Incorrect col:", x_col, y_col)
```

```python
def nameOfCol(col):
    cols_list = ["year", "region", "npg", "birth_rate", "death_rate", "gdw", "urbanization"]
    if 6 < col or col < 0:
        return -1
    return cols_list[col]

dataset = pd.read_csv('for4lab.csv')
dataset = dataset.dropna()

if nameOfCol(x_col) == -1 or nameOfCol(y_col) == -1:
    print("Can't get cols from file")
else:
    # work with dataset
    x = dataset[nameOfCol(x_col)]
    y = dataset[nameOfCol(y_col)]
    
    linear_regression = LinearRegression()
    
    # reshape
    x = np.array(x).reshape(-1, 1)
    y = np.array(y).reshape(-1, 1)
    
    # fit data
    linear_regression.fit(x, y)
    y_pred1 = linear_regression.predict(x)
    plt.scatter(x, y)
    plt.xlabel(nameOfCol(x_col))
    plt.ylabel(nameOfCol(y_col))
    plt.plot(x, y_pred1, color="red")
    plt.show()
```

```python
x = int(input("input x: "))
x = np.array(x).reshape(-1, 1)
y = linear_regression.predict(x)
print("predict: ", y[0][0])
```

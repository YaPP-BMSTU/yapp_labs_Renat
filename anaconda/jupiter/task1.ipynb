```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import warnings
```

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/3/38/Jupyter_logo.svg/1200px-Jupyter_logo.svg.png" alt="картинка">

```python
cleaned_ships_data = pd.read_csv('Cleaned_ships_data.csv')
```

<h3>2-D график зависимости средней максимальной загрузки кораблей от длины корабля<h3/>

```python
lineplot = sb.lineplot(cleaned_ships_data, x = "dwt", y = "length").set_title('График зависимости средней маскимальной загрузки кораблей от длины корабля\n')
xlabel = plt.xlabel('\nсредняя максимальная загрузка корабля')
ylabel = plt.ylabel('длина корабля\n')
```

```python
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
```

___

<h3>Распределение количества кораблей по их типу (ship_name).<h3/>

```python
ships = cleaned_ships_data.groupby("ship_name").sum('gt')['gt']
names = cleaned_ships_data["ship_name"].unique().tolist()
plt.pie(ships.tolist(), labels=names, autopct='%1.1f%%')
plt.title('Распределение количества кораблей по их типу')
plt.show()
```

___

<h3>Гистограмма распределения количества кораблей по годам постройки (built_year)<h3/>

```python
sb.displot(cleaned_ships_data, x = "built_year")
plt.title('Гистограмма распределения количества кораблей по годам постройки\n')
xlabel_dp = plt.xlabel('\nгод постройки')
ylabel_dp = plt.ylabel('количество\n')
```

___

<h3>2-D Гистограмма распределения количества кораблей по годам постройки (built_year) и максимальному весу груза (dwt). Для максимального веса груза сформировать 20 диапазонов, от минимального до максимального значений с равным шагом.<h3/>

```python
iris = pd.read_csv('IRIS.csv')
sb.pairplot(iris)
```

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import warnings
```

```python
cleaned_ships_data = pd.read_csv('Cleaned_ships_data.csv')
```

<h3>2-D график зависимости средней максимальной загрузки кораблей от длины корабля<h3/>

```python
plt.figure(figsize=(15,7))
lineplot = sb.lineplot(cleaned_ships_data, y = "dwt", hue = "ship_name", x = "length").set_title('График зависимости средней маскимальной загрузки кораблей от длины корабля\n')
xlabel = plt.xlabel('\nсредняя максимальная загрузка корабля(кг)')
ylabel = plt.ylabel('длина корабля(м)\n')
```

```python
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
```

___

<h3>Распределение количества кораблей по их типу (ship_name).<h3/>

```python
ship_counts = cleaned_ships_data['ship_name'].value_counts()
ship_names = ship_counts.index
ship_counts_values = ship_counts.values
plt.figure(figsize=(8, 8))
plt.bar(ship_names, ship_counts_values)
plt.xlabel('Тип корабля')
plt.ylabel('Количество кораблей')
plt.title('Распределение количества кораблей по их типу')
plt.xticks(rotation=90)
plt.show()
```

___

<h3>Гистограмма распределения количества кораблей по годам постройки (built_year)<h3/>

```python
sb.displot(cleaned_ships_data, x = "built_year")
plt.title('Гистограмма распределения количества кораблей по годам постройки\n')
xlabel_dp = plt.xlabel('\nгод постройки')
ylabel_dp = plt.ylabel('количество\n')
```

___

<h3>2-D Гистограмма распределения количества кораблей по годам постройки (built_year) и максимальному весу груза (dwt). Для максимального веса груза сформировать 20 диапазонов, от минимального до максимального значений с равным шагом.<h3/>

```python
dwt_ranges = np.linspace(cleaned_ships_data['dwt'].min(), cleaned_ships_data['dwt'].max(), num = 21)
plt.figure(figsize=(12,6))
plt.hist2d(cleaned_ships_data['built_year'], cleaned_ships_data['dwt'], bins = [20, dwt_ranges], cmap='Greens')
plt.title('Гистограмма распределения количества кораблей по годам постройки и максимальному весу груза\n')
plt.colorbar(label='\nколичество кораблей')
plt.xlabel('\nгоды постройки')
plt.ylabel('максимальный вес груза, кг\n')
plt.show()
```

```python
iris = pd.read_csv('IRIS.csv')
pp = sb.pairplot(iris, hue='species')
```

<h3>Парную диаграмму (Pairplot) для датасета<h3/>

<h3>Скрипичную диаграмму (Violinplot) с распределением всех четырех характеристик ирисов для каждого вида (для каждой характеристики отдельные “скрипки” для разных видов ирисов)<h3/>

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sb.violinplot(x="species", y="sepal_length", data=iris)
plt.ylabel('длина чашелистника, см')
plt.subplot(2, 2, 2)
sb.violinplot(x="species", y="sepal_width", data=iris)
plt.ylabel('ширина чашелистника, см')
plt.subplot(2, 2, 3)
sb.violinplot(x="species", y="petal_length", data=iris)
plt.ylabel('длина лепестка, см')
plt.subplot(2, 2, 4)
sb.violinplot(x="species", y="petal_width", data=iris)
plt.ylabel('ширина лепестка, см')
plt.tight_layout()
plt.show()
```
